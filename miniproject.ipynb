```python
## [1-1] 모듈 로딩
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = 'C:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name) 
```

```python
## [1-2] 데이터 준비
D_FILE = './고독사_발생_현황_지역별_20250201173217.xlsx'
```

```python
## [1-3] 데이터 로딩
loneDF=pd.read_excel(D_FILE)
loneDF
```

```python
loneDF.info()
```

```python
loneDF = loneDF.iloc[1:].reset_index(drop=True)
loneDF
```

```python
## 데이터 시각화
plt.figure(figsize=(20, 10))
plt.title("연도별 고독사 추이")


# 레이블 리스트 (컬럼명과 일치해야 함)
labels = ['2017','2018','2019','2020','2021']

# 색상 리스트
colors = ['#ffd4e5', '#d4ffea', '#eecbff', '#feffa3', '#dbdcff', '#f9a7a7']

# 막대 너비 설정
width = 0.15

# X축 (시도별)
x = loneDF.index.astype(int)  # 인덱스를 정수형으로 변환
# Y축 (고독자수)
y = loneDF.index.astype(int)

# 막대 그래프 생성
for i, (label, color) in enumerate(zip(labels, colors)):
    plt.bar(x + (i - len(labels) / 2) * width, loneDF[label].astype(float), 
            width=width, label=label, color=color)  # 컬럼명을 직접 사용

plt.legend(title="연도", fontsize=14)  # 범례 추가
plt.xlabel("시도별" , fontsize=14)
plt.ylabel("고독자 수" , fontsize=14)
plt.xticks(x, loneDF['시도별'], fontsize=10)  # X축 눈금 설정
plt.yticks(np.arange(0,3500, 100), fontsize=10)  # Y축 눈금 설정 (0~2500, 100 단위)
plt.grid(axis='y', linestyle='--', alpha=1.0)  # Y축 격자선 추가
plt.show()
```
